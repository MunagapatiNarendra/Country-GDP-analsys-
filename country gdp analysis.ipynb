```python
import matplotlib.pyplot as plt #visulaization
import seaborn as sns #distribution visulaization
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,5

#import warnings 
#warnings.filterwarnings('ignore')
```

```python
df.head()
```

```python
#Distribution:
vis1 = sns.displot(df["InternetUsers"])
```

```python
vis1 = sns.distplot(df["InternetUsers"], bins=10)
```

```python
#BOX PLOTS:
vis2 = sns.boxplot(data = df, x="IncomeGroup", y='BirthRate')
```

```python
#refer to seaborn gallary
#visualizing with seaborn
```

```python
vis3 = sns.lmplot(data = df, x='InternetUsers', y='BirthRate')#this is called distribution line
```

```python
#hue - parameter for color
vis4 = sns.lmplot(data = df, x="InternetUsers",y="BirthRate",
                   fit_reg = True, hue = "IncomeGroup")
```

```python
vis5 = sns.lineplot(data = df,x="InternetUsers",y="BirthRate")
```

```python
vis6 = sns.violinplot(data = df,x="InternetUsers",y="BirthRate") 
```
